# EDA Report 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Applicant.xlsx')
df.head()
```

## Basic Dataset Overview

```python
# Shape of dataset
print("Rows and Columns:", df.shape)

# Column names
print("\nColumns:\n", df.columns)

# Data types and non-null counts
df.info()
```

## Summary Statistics

```python
# Numerical Summary
df.describe()
```

```python
# Categorical summary
df.describe(include='object')
```

## Missing Value Analysis

```python
# Missing value count
missing_count = df.isnull().sum()

# Missing value percentage
missing_percentage = (df.isnull().mean() * 100).round(2)

missing_df = pd.DataFrame({
    'Missing Count': missing_count,
    'Missing Percentage (%)': missing_percentage
})

missing_df.sort_values(by='Missing Percentage (%)', ascending=False)
```

## Duplicate Record Check

```python
# Total duplicate rows
duplicates = df.duplicated().sum()
print("Total Duplicate Rows:", duplicates)
```

```python
# Check uniqueness of Reference_ID
df['Reference_ID'].nunique()
```

## Convert Timestamp Columns to Datetime

```python
timestamp_cols = ['Recieved_At', 'Created_At', 'Modified_At', 'Admit_Date']

for col in timestamp_cols:
    df[col] = pd.to_datetime(df[col], unit='ms', errors='coerce')

df[timestamp_cols].head()
```

## Univariate Analysis – Categorical Variables

### Country Distribution

```python
df['Country'].value_counts()
```

```python
plt.figure(figsize=(6,4))
df['Country'].value_counts().plot(kind='bar')
plt.title("Applicant Distribution by Country")
plt.xlabel("Country")
plt.ylabel("Number of Applicants")
plt.show()
```

### Intake Distribution

```python
plt.figure(figsize=(6,4))
df['Intake'].value_counts().plot(kind='bar')
plt.title("Applicants by Intake")
plt.xlabel("Intake")
plt.ylabel("Count")
plt.show()
```

## Top Majors Applied For

```python
df['Major'].value_counts().head(10)
```

```python
plt.figure(figsize=(8,4))
df['Major'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Majors Applied")
plt.xlabel("Major")
plt.ylabel("Applicants")
plt.show()
```

## Time-Based Analysis

```python
df['Application_Month'] = df['Recieved_At'].dt.to_period('M')
df['Application_Month'].value_counts().sort_index()
```

```python
df['Application_Month'].value_counts().sort_index().plot(figsize=(10,4))
plt.title("Applications Over Time")
plt.xlabel("Month")
plt.ylabel("Number of Applications")
plt.show()
```

# Bivariate Analysis

```python
pd.crosstab(df['Country'], df['Intake'])
```

```python
pd.crosstab(df['Country'], df['Intake']).plot(
    kind='bar', figsize=(8,4)
)
plt.title("Country vs Intake")
plt.xlabel("Country")
plt.ylabel("Applicants")
plt.show()
```
